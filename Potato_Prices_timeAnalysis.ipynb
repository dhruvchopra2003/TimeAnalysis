```python
df_new = df[['ds','y']]
df_new
```

```python
df_new['ds']= pd.to_datetime(df_new['ds']) #to convert object to datetime format that is acceptable by pandas
df_new['y']= df_new['y'].astype(int)# type casting from object to integer
```

```python
df_new.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(25,10))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
import numpy as np
df_new['y'] = np.where(df_new['y']>2500,np.nan, df_new['y'])
df_new
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(25,10))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
#FBPROPHET
import fbprophet
op = fbprophet.Prophet()
```

```python
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods=365, freq='D')
```

```python
forecast=op.predict(forecast)
forecast  
```

```python
plt.figure(figsize =(20,10))
op.plot(forecast, xlabel='Date', ylabel = 'potato price/q')
plt.show()
```
